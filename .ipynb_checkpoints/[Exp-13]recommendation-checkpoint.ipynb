```python
# 모든 영어 문자는 소문자로 만들고, 섞여있는 html 태그를 제거하고,
# 정규 표현식을 통해 각종 특수문자를 제거한다.

# 데이터 전처리 함수
def preprocess_sentence(sentence, remove_stopwords=True):
    sentence = sentence.lower() # 텍스트 소문자화
    sentence = BeautifulSoup(sentence, "lxml").text # <br />, <a href = ...> 등의 html 태그 제거
    sentence = re.sub(r'\([^)]*\)', '', sentence) # 괄호로 닫힌 문자열 (...) 제거 Ex) my husband (and myself!) for => my husband for
    sentence = re.sub('"','', sentence) # 쌍따옴표 " 제거
    sentence = ' '.join([contractions[t] if t in contractions else t for t in sentence.split(" ")]) # 약어 정규화
    sentence = re.sub(r"'s\b","", sentence) # 소유격 제거. Ex) roland's -> roland
    sentence = re.sub("[^a-zA-Z]", " ", sentence) # 영어 외 문자(숫자, 특수문자 등) 공백으로 변환
    sentence = re.sub('[m]{2,}', 'mm', sentence) # m이 3개 이상이면 2개로 변경. Ex) ummmmmmm yeah -> umm yeah
    
    # 불용어 제거 (Text)
    if remove_stopwords:
        tokens = ' '.join(word for word in sentence.split() if not word in stopwords.words('english') if len(word) > 1)
    # 불용어 미제거 (headlines)
    else:
        tokens = ' '.join(word for word in sentence.split() if len(word) > 1)
    return tokens
print('=3')
```

## 1.4. 이상치 제거

기존 예시에서와 다르게, 이상치를 특별히 제거할 필요는 없다고 여겨졌습니다.
text와 headlines 모두 아주 예쁜 모양의 정규분포를 그리고 있었습니다.
하지만, 그래도 이상치는 제거하는 것이 맞다고 보여져서, text와 headlines의 분포 그래프를 시각적으로 확인하고 manual하게 상한(upper bound)과 하한(lower bound)을 정했습니다.

```python
# 길이 분포 출력
import matplotlib.pyplot as plt

text_len = [len(s.split()) for s in data['text']]
headlines_len = [len(s.split()) for s in data['headlines']]

print('텍스트의 최소 길이 : {}'.format(np.min(text_len)))
print('텍스트의 최대 길이 : {}'.format(np.max(text_len)))
print('텍스트의 평균 길이 : {}'.format(np.mean(text_len)))
print('헤드라인의 최소 길이 : {}'.format(np.min(headlines_len)))
print('헤드라인의 최대 길이 : {}'.format(np.max(headlines_len)))
print('헤드라인의 평균 길이 : {}'.format(np.mean(headlines_len)))

plt.subplot(1,2,1)
plt.boxplot(text_len)
plt.title('Text')
plt.subplot(1,2,2)
plt.boxplot(headlines_len)
plt.title('Headlines')
plt.tight_layout()
plt.show()

plt.title('Text')
sns.histplot(text_len, binwidth=1)
plt.xlabel('length of samples')
plt.ylabel('number of samples')
plt.show()

plt.title('Headlines')
sns.histplot(headlines_len, binwidth=1)
plt.xlabel('length of samples')
plt.ylabel('number of samples')
plt.show()
```

```python
text_min_len = 20
text_max_len = 50
headlines_min_len = 5
headlines_max_len = 15
print('=3')
```

```python
# 훈련데이터(nested_list)와 샘플의 길이(max_len)를 입력했을 때, 데이터의 몇 %가 max_len보다 적은지를 출력하는 함수
def below_threshold_len(min_len, max_len, nested_list):
  cnt = 0
  for s in nested_list:
    if((len(s.split()) <= max_len)&(len(s.split()) >= min_len)):
        cnt = cnt + 1
  print('전체 샘플 중 길이가 적정 범위 안에 있는 샘플의 비율: %s'%( (cnt / len(nested_list))))
print('=3')
```

```python
# len(data['Text'][0].split())
```

```python
below_threshold_len(text_min_len, text_max_len, data['text'])
below_threshold_len(headlines_min_len, headlines_max_len,  data['headlines'])
```

```python
print("text의 길이가 적정 범위 안에 있는 sample의 개수 : {}".format(
np.sum((np.array([len(text.split()) for text in data['text']]) > text_max_len)
       |(np.array([len(text.split()) for text in data['text']]) < text_min_len))
))
print("headlines의 길이가 적정 범위 안에 있는 sample의 개수 : {}".format(
np.sum((np.array([len(text.split()) for text in data['headlines']]) > headlines_max_len)
       |(np.array([len(text.split()) for text in data['headlines']]) < headlines_min_len))
))
```

```python
# data['Text']의 원소 중에서 단어의 개수가 text_max_len = 50보다 적은 문장들만
# 남긴다. 위의 주석처리한 코드 A. 이것은 단순히 series 단위에서 삭제하는 것이
# 아니라, DataFrame 단위에서 행을 삭제한 것이다. 위의 주석처리한 코드 B
# data['headlines']에서도 마찬가지이다.
data = data[data['text'].apply(lambda x: len(x.split()) <= text_max_len)]
data = data[data['text'].apply(lambda x: len(x.split()) >= text_min_len)]
data = data[data['headlines'].apply(lambda x: len(x.split()) <= headlines_max_len)]
data = data[data['headlines'].apply(lambda x: len(x.split()) >= headlines_min_len)]
print('전체 샘플수 :', (len(data)))
```

## 1.6. 데이터 토큰화

### 1.6.1. train set 토큰화

```python
# train 데이터의 단어들을 모두 정수로 만들어준다. 즉, 각 문장들을 euclidean vector로 만들어준다. 이 과정을 '토큰화'라고도 했었다.

src_tokenizer = Tokenizer() # 토크나이저 정의
src_tokenizer.fit_on_texts(encoder_input_train) # 입력된 데이터로부터 단어 집합 생성
print('=3')
```

```python
# 단어장을 만들었다. 단어장에 등록된 단어의 수는 총 32017개이다.
# src_tokenizer.word_index
# len(src_tokenizer.word_index)
```

```python
# 단어들이 너무 많다. 아니, 그보다는, 학습의 용이성을 위해 자주 등장하지 않는 단어들을 배제시키려고 한다.
# 아래 코드를 통해 보면, 일곱 번 이상 등장하는 단어들의 개수는 약 8000개이다. 이 8000개만 단어장에 등록할 것이고,
# 나머지 단어들은 out of vocabulary로 처리하려나 보다.
threshold = 7
total_cnt = len(src_tokenizer.word_index) # 단어의 수
rare_cnt = 0 # 등장 빈도수가 threshold보다 작은 단어의 개수를 카운트
total_freq = 0 # 훈련 데이터의 전체 단어 빈도수 총 합
rare_freq = 0 # 등장 빈도수가 threshold보다 작은 단어의 등장 빈도수의 총 합

# 단어와 빈도수의 쌍(pair)을 key와 value로 받는다.
for key, value in src_tokenizer.word_counts.items():
    total_freq = total_freq + value

    # 단어의 등장 빈도수가 threshold보다 작으면
    if(value < threshold):
        rare_cnt = rare_cnt + 1
        rare_freq = rare_freq + value

print('단어 집합(vocabulary)의 크기 :', total_cnt)
print('등장 빈도가 %s번 이하인 희귀 단어의 수: %s'%(threshold - 1, rare_cnt))
print('단어 집합에서 희귀 단어를 제외시킬 경우의 단어 집합의 크기 %s'%(total_cnt - rare_cnt))
print("단어 집합에서 희귀 단어의 비율:", (rare_cnt / total_cnt)*100)
print("전체 등장 빈도에서 희귀 단어 등장 빈도 비율:", (rare_freq / total_freq)*100)
```

```python
# 아까 말한대로, 총 단어의 수를 8000개로 제한시켰다. 이후에 나오는 코드를 보면 Tokenizer라는 메서드(?)를 쓴 것은 똑같지만
# num_words를 8000으로 특정했다는 점만 달라졌다. 그렇게 해서, 새로운 tokenizer를 만들었다.
src_vocab = 8000
src_tokenizer = Tokenizer(num_words=src_vocab) # 단어 집합의 크기를 8,000으로 제한
src_tokenizer.fit_on_texts(encoder_input_train) # 단어 집합 재생성

# 텍스트 시퀀스를 정수 시퀀스로 변환
encoder_input_train = src_tokenizer.texts_to_sequences(encoder_input_train) 
encoder_input_test = src_tokenizer.texts_to_sequences(encoder_input_test)

# 잘 진행되었는지 샘플 출력
print(encoder_input_train[:3])
```

# 2. 모델 구성 및 학습

이번 exploration node에서 구성한 코드는, 다음과 같은 모델에 대한 코드이다.

## 2.1. vanilla seq2seq model

궁극적으로 구성하고자 하는 모델은, 여기 (**2.1.**)에서 구성하는 vanilla seq2seq 모델은 아니다.
그리고 Bahdanau-style의 모델도 (bidrectinal이 아니고 stacked되었다는 점에서) 엄밀히 말하면 아니다.
하지만 2.1.에서 설명하고자 하는 vanilla seq2seq 모델과 전체적인 구조는 유사하므로, 일단 vanilla seq2seq 모델을 설명할 것 같다.
그리고, 추후에 2.2.에서 이 코드의 모델을 설명할 때에 많은 부분은 생략될 것 같다.
여기의 설명으로 충분할 것 같기 때문이다.

위 그림에 나타난 text는 "new zealand ... history"라는 문장이고, 이것은 encoder의 input으로 들어간다.
그리고 headline은 "new zealand ... win first match ... cricket history"라는 문장이다.
이 문장의 맨 앞에 sos를 붙인 문장을 decoder의 input으로 넣고, 이 문장의 맨 뒤에 eos를 붙인 문장을 decoder의 label로, 즉 이 모델의 label로 삼는다.

다시 말해서, 왼쪽의 파랑색이 포함된 부분이 encoder이고, 오른쪽에 빨강색과 갈색이 포함된 부분이 decoder이다.

위의 그림에서 'new', 'zealand', 'history'와 같은 단어들은 embedding layer(주황색 직사각형)을 거쳐서 word vector로 출력된다.
여기에서 word vector는 128차원의 벡터이다.
오른쪽 아래에 있는 주황색 직사각형도 embedding layer이다.
다만, 오른쪽 위에 있는 주황색 직사각형은 embedding layer라기보다는, word vector를 단어로 다시 복원해주는 layer가 되겠다.

encoder의 n_timesteps (the number of timesteps)는 그림상에서 N으로 표시되어 있다.
decoder의 경우 n_timesteps가 T이다.
이 코드 상에서는 N=50이고 T=16으로 구현되어 있다.

encoder의 timestep은 i로 표시할 것이고, decoder의 timestep은 t로 표시할 것이다.
i, t의 상한은 각각 N, T로 잡았다. ($1\le i\le N$, $1\le t\le T$)

encoder 그림에서 보이는 파랑색 구조는 세 번 쌓인 stacked LSTM 구조를 보여준다.
그래서 처음 두 번은 `return_sequences=True`로 놓는 것이 당연한데, 세번째도 `return_sequences=True`로 놓는다.
따라서 이 encoder는 many-to-many의 구조(이면서 input과 output의 timestep이 정확히 일치하는 구조)이다.
그리고 세 LSTM에서 모두 `return_state=True`로 놓았다. 이것은 hidden state뿐만 아니라 cell state도 return하겠다는 뜻이다.
hidden state과 cell state 모두 256차원의 벡터이다.

여기서부터는 vanilla seq2seq만의 특징이다.
encoder의 세번째 LSTM 레이어에서의 hidden state와 cell state는 decoder의 initial hidden state와 initial cell state로 들어간다.
그리곤 입력받은 $u_i$들과 함께 LSTM을 구성한다.
$u_i$들은 128차원의 word vector인데, 이것들이 decoder의 hidden state과 cell state와 함께 새로운 hidden state와 cell state를 만들어낸다 (LSTM에 대한 설명이다.) 그렇게 하여 얻은 hidden state들은 복원레이어 (word vector를 다시 단어로 복원해주는 layer, 이러한 함수, 즉 word embedding의 정확한 역함수는 존재하지 않을 것이지만, 그런 함수가 있다고 하자. 아마 코드의 뒷부분을 열심히 읽으면 되겠지만, 거기까지 자세히 읽지는 않았다.)를 거쳐 다시 단어를 출력해주는 구조가 된다.

```python
# encoder_inputs
# enc_emb
# encoder_output1, state_h1, state_c1
# encoder_output2, state_h2, state_c2
```

```python
# encoder를 설계한다.
# embedding_dim은 128로 두었다. 이것은 word vector의 dimension이다.
# hidden_size는 256으로 두었다. 이것은 hidden state vector의 dimension이다.
# 앞에서 hidden state 말고 cell state도 고려한다고 했었다. 
#
# Input : Input() is used to instantiate a Keras tensor.
# encoder_input은 shape이 (None, 50)인 keras tensor이다. 이때 50은 (패딩 이후) 각 문장의 단어 수이다. 즉 n_timesteps이다.
#
# Embedding : Turns positive integers (indexes) into dense vectors of fixed size.
# enc_emb는 shape이 (None, 50, 128)인 keras tensor이다. 128은 embedding_dim이다.
# enc_emb는 encoder_input을 embedding시킨 벡터들로 구성되어 있으며 LSTM의 input에 해당하기도 한다.
#
# 일반적으로 LSTM input은 rank 2 tensor이거나 (n_timesteps, n_features) rank 3 tensor이다. (n_batchs, n_timesteps_, n_features)
# 여기에서는 후자를 따르는 것 같다. 참고로 n_feature에서 이것이 꼭 "feature의 개수"를
# 의미한다고는 볼 수 없다. 이 예시에서는 n_features=50인데, 이것은 각 timestep에 들어가는 벡터의 dimension이고
# stacked LSTM에서 두번째 층의 LSTM의 입장에서는, 첫번째 LSTM의 hidden state의 dimension일 것이다.
# enc_emb를 LSTM에 넣는다. 이 LSTM은 stacked LSTM이기 때문에 return_sequences=True로 놓는다.
# return_state=True는 처음 보는건데, 찾아보니 hidden state vector 말고도 cell state vector도 return한다는 의미인 것 같다.
# 개념적으로, hidden state와 cell state의 dimension은 같을 수밖에 없다. 따라서 cell state의 dimension을 따로 명시하지 않아도 된다.
#
# encoder_output1은 shape이 (None, 50, 256)인 keras tensor이다.
# 이것은 단지 아래 설명할 hidden state가 50개 겹쳐져 있는 것을 한번에 표현한 것뿐이다.
# state_h1은 shape이 (None, 256)인 keras tensor이다. hidden state vector를 의미한다. 이때, None은 나중에 설정할 batch size에 해당한다.
# state_c1은 shape이 (None, 256)인 keras tensor이다. cell state vector를 의미한다.

from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Concatenate, TimeDistributed
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# 인코더 설계 시작
embedding_dim = 128
hidden_size = 256

# 인코더
encoder_inputs = Input(shape=(text_max_len,))

# 인코더의 임베딩 층
enc_emb = Embedding(src_vocab, embedding_dim)(encoder_inputs)

# 인코더의 LSTM 1
# encoder_lstm1 = LSTM(hidden_size, return_sequences=True, return_state=True ,dropout = 0.4, recurrent_dropout = 0.4)
encoder_lstm1 = LSTM(hidden_size, return_sequences=True, return_state=True ,dropout = 0.4)
encoder_output1, state_h1, state_c1 = encoder_lstm1(enc_emb)

# 인코더의 LSTM 2
# encoder_lstm2 = LSTM(hidden_size, return_sequences=True, return_state=True, dropout=0.4, recurrent_dropout=0.4)
encoder_lstm2 = LSTM(hidden_size, return_sequences=True, return_state=True, dropout=0.4)
encoder_output2, state_h2, state_c2 = encoder_lstm2(encoder_output1)

# 인코더의 LSTM 3
# encoder_lstm3 = LSTM(hidden_size, return_state=True, return_sequences=True, dropout=0.4, recurrent_dropout=0.4)
encoder_lstm3 = LSTM(hidden_size, return_state=True, return_sequences=True, dropout=0.4)
encoder_outputs, state_h, state_c= encoder_lstm3(encoder_output2)
```

```python
# 이 모든 것은 functional api를 통해 구성되었다고 한다.
# 그것이 뭔지 까먹었거나, 잘 알지 못하지만 어쨌든 기존처럼 sequential하게 만들지 않은 것만은 분명하다.
# 처음에 encoder_inputs라는 객체에 입력 shape을 결정해준 듯하고 (batch_size가 무엇이든지 간에 수용할 수 있도록 해준 것이 아닐까 한다.)
# 최종적으로는 encoder_outputs를 얻었다. 그 외에 state_h와 state_c도 얻었지만, state_h는 기존의 encoder_outputs에 중복된다.
# state_c는 어떻게 쓰이는지 다시 봐야 할 것이다.
#
# 다음으로는 디코더를 설계한다.
```

```python
# encoder에 비해서 decoder는 상대적으로 간단하다. 그리고 아까와 비슷하기도 하다.
# decoder_input은 shape이 (None, None)인 keras tensor이다. 그러니까 rank=2인 것만 주어졌지 다른 건 주어지지 않았다.
# 하지만 아까의 문맥에 대입해 생각해보면, 처음 None은 batch size를, 뒤의 None은 n_timesteps가 되어야 한다.
# 그런데 50이라고 표시되지는 않고 있다.
#
# dec_emb는 shape이 (None, None, 128)인 keras tensor이다. 마찬가지의 문맥에서 보면 맨앞의 None은 batch size,
# 두번째 None은 n_timesteps=8일 것이다. 128은 word vector의 dimension이라고 할 수 있고, 이것은 encoder에서와 같다.
#
# enc_emb와 마찬가지로 dec_emb는 LSTM의 input이다.
# 이번에는 하나의 LSTM만 쌓는다. 이때 이 LSTM의 hidden state와 cell state를, 직전 LSTM의 hidden state와 cell state로 initialize해준다.
# 그것이 initial_state 인자인 것 같다. 그것은 tensorflow api에도 명확히 표현되어 있지 않은 것 같다. 다음과 같이 표현되어 있다.
# initial_state : List of initial state tensors to be passed to the first call of the cell
# (optional, defaults to None which causes creation of zero-filled initial state tensors).
#
# 어쨌든, Aiffel node 상에 그렇게 표현되어 있으니 맞을 것 같다.
# decoder도 return_sequences=True로 놓았다. 다시 말해 timestep이 8인 tensor를 다시 뱉어내게 되어 있다.
# 이것은 맞는 말이다. 우리가 해야 할 것은 문장을 만들어내는 것이다. 그 문장이 단어 8개로 구성될 필요까지는 없겠지만,
# 여하튼 최대 8개 단어로 구성된 Summary를 만드는 게 우리의 목적이다.
# 최종적인 결과로 나오는 것은 decoder_outputs인데, 이것은 (None, None, 256) 형태의 rank 3 tensor이다.
# 중간의 None을 8로 간주하면 (batch size만큼의) timestep=8의 256차원 벡터들을 의미한다.

# 디코더 설계
decoder_inputs = Input(shape=(None,))

# 디코더의 임베딩 층
dec_emb_layer = Embedding(tar_vocab, embedding_dim)
dec_emb = dec_emb_layer(decoder_inputs)

# 디코더의 LSTM
# decoder_lstm = LSTM(hidden_size, return_sequences=True, return_state=True, dropout=0.4, recurrent_dropout=0.2)
decoder_lstm = LSTM(hidden_size, return_sequences=True, return_state=True, dropout=0.4)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=[state_h, state_c])
```

```python
# 하지만 아직 끝나지 않았다. 위에 설명한 대로, decoder_outputs는 단지 256차원의 벡터들의 8개 모임(의 모임)일 뿐이다.
# 그러나, 우리가 원하는 건, tar_vocab=2000차원의 벡터들의 8개 모임(의 모임)을 원한다.
# 이를 위해, tar_vocab=2000개의 노드로 구성된 fully connected 레이어를 구성하는데 activation function을 softmax로 둔다.
# 
# 그리고 드디어, 모델을 정의한다. 잘 모르겠지만, functional-api로 모델을 정의할 때에는, 입력값 x와 그에 대한 출력값 \hat y를 명시해주면
# 그것으로 함수(즉 모델)이 구성되는 것 같다.
# model.summary()로 모델을 한눈에 볼 수 있다.
# 복습하듯 다시 보면, shape들에서 맨 앞의 None은 항상 batch size를 의미한다.
# InputLayer에서 (최대) 길이가 50인 문장을 입력받고 (encoder의 input) 이것은 다시 embedding layer를 거쳐 128차원 벡터들의
# (최대 50개) 모임으로 변환되고 (word embedding) 이것이 그대로 3 times stacked LSTM의 입력으로 들어간다.
# LSTM의 hidden state와 cell state은 256차원의 벡터들이다.
# 마지막 LSTM의 hidden state와 cell state는 decoder의 hidden state와 cell state의 initial value로 들어간다.
# 한편 decoder에도 (최대) 길이가 50인 문장을 입력받고 

# 디코더의 출력층
decoder_softmax_layer = Dense(tar_vocab, activation='softmax')
decoder_softmax_outputs = decoder_softmax_layer(decoder_outputs) 

# 모델 정의
model = Model([encoder_inputs, decoder_inputs], decoder_softmax_outputs)
model.summary()
```

```python
# 여기까지는 (아마도) 기존의 seq2seq 모델이었던 것 같다. 그리고 새로운 내용인 attention이 추가된다.
# 사용하는 attention의 방식은 Bahdanau 스타일의 attention이라고 한다.
# tensorflow api의 AdditiveAttention에는 다음과 같은 짤막한 설명이 있다 : Additive attention layer, a.k.a. Bahdanau-style attention.
# 이 Bahdanau 알고리즘에 대해서는 이 글 정도만 읽어보았다. https://machinelearningmastery.com/the-bahdanau-attention-mechanism/
# 다 이해하지는 못했고, 이해한 사실 몇가지는 다음과 같다.
#
# 1. 기존의 접근법에서는 (GRU를 만든 조경현 교수님도 언급되고 있다.) RNN encoder-decoder 방식의 아키텍쳐를 구성하고 있기는 하지만
# input이 variable length sentence인데 비해 context vector가 fixed length vector라는 것이 문제점이라고 지적하고 있다.
# Bahdanau 방식은 이 방법을 해결하는데, input sentence를 a sequence of vectors로 바꾼 후 그것들의 일부(subset)을 적절히 골라서
# decode하는데 사용한다고 되어 있다.
# 원문 :Instead, it encodes the input sentence into a sequence of vectors and chooses a subset of these vectors adaptively
# while decoding the translation.
#
# 2. Bidrectional RNN을 사용하는데, encoder와 decoder 모두에서 사용한다.
#
# 3. s_t, h_i, c_t, alpha_{t,i}, e_{t,i}
# s, c, h를 지정할 때 두 종류의 index가 있다 : i, t
# 아마도 i는 encoder에서의 timestep을, t는 decoder에서의 timestep을 의미한다.
# i에 대해서는 더 할말이 없고, t에 대해서는 할 말이 있다. 즉, 각각의 i에 대하여 context vector c_t와 attention weights을 전달받아서
# 적당히 weighted sum을 적용하여 y_t를 만드는 것 같다. 구체적으로는 다음과 같다.
# h_i는 encoder 단계의 hidden state vector인데, bidirectional RNN이라는 점에서 서로 다른 두 방향의 hidden state을 concat한 것이다.
# encoder를 먼저 수행하므로, 모든 i에 대하여 h_i가 계산되어 있다고 가정할 수 있다. 그런 상황에서 decoder의 timestep t가 되었다고 해보자.
# 그리고 s_1, s_2, ..., s_{t-1}까지를 알고 있는 상태에서 s_t를 구하는 것이 목적이다.
# 고정된 t와 이미 알고있는 모든 i에 대하여, e_{t,i}를 계산한다. 그것을 softmax로 normalize하여 \alpha_{t,i}를 얻는다.
# 이때 e_{t,i}를 계산하는 과정에서 addtion operation이 포함된다는 점에서 이 알고리즘을 AdditiveAttention이라고 부른다.
# 구체적으로는 (모든 i에 대하여) e_{t,i}=a(s_{t-1},h_i)로 표현되는데 h_i과 s_{t-1}을 concat하고, weight matrix W를 곱한 후
# hyperbolic tangent를 취하고 적당한 weight vector v를 내적해줌으로써 얻어진다. W와 v는 learnable parameter인 듯하다.
# 그렇다면, s_{t-1}와 h_i를 더하는 것이 아니라 concatenate하는 것인데, concatenation을 addition으로 본 것 같다.
#
# 4. 다시 말해 이 attention mechanism에서 quary가 뭐고 key가 뭔지는 잘 모르겠다.
# 하지만 이렇게 얻은 \alpha_{t,i}들을 h_i들을 가지고 weighted sum을 만든다는 점에서 h_i가 value vector의 역할을 한다는 것은 분명해보인다.
# 그렇게 얻은 weighted sum을 c_t라고 부른다. 이것이 context vector가 되는 것이다. 
# 이 context vector는 t에 depend하는 값이다. 즉, decoder의 각 timestep t마다 업데이트 되는 벡터라는 뜻이다.
# 이는 exploration의 설명과도 어느 정도 일치한다.
#
# 여하튼 그렇게 하여, 모델을 설계하게 된다.
```

## 2.2. seq2seq model with attention

vanilla seq2seq의 대표적인 특징이자 단점은 context vector가 hidden state vector라는 것이다.
Bahdanau는 그의 논문에서, 이러한 vanilly seq2seq의 단점에 대하여 다음과 같이 지적했다.
([Neural Machine Translation by Jointly Learning to Align and Translate](https://arxiv.org/abs/1409.0473), 2014)

> The most important distinguishing feature of this approach from the basic encoder-decoder is that it does not attempt to encode a whole input sentence into a single fixed-length vector. Instead, it encodes the input sentence into a sequence of vectors and chooses a subset of these vectors adaptively while decoding the translation.

'a sequence of vectors'의 의미는 여기에서 'a sequence of hidden state vectors'가 되겠고, 'a subset of these vectors'만을 adaptively하게 선택한다는 것은 hidden state vector들에 대하여 가중치를 두어 합하겠다는 뜻이다.
구체적인 "seq2seq model with attention"에 대한 설명은 다음과 같다.

여기에서는 이 코드의 구조에 맞게 설명하려고 한다.
하지만 Bahdanau 논문에 나온 모델의 구조와 거의 같다.
다른 점이라면 다음의 두 가지이다.
 - Bahdanau 모델에서는 encoder에 LSTM 한 개를 쓴 것 같은데, 여기에서는 세 개가 쌓인 stacked LSTM을 사용했다.
 - Bahdanau 모델에서는 bidrectional LSTM을 사용한 것 같은데, 여기에서는 그냥 LSTM을 사용했다.
 
지금 하려고 하는 설명은, Bahdanau의 논문을 직접 공부해서 얻은 설명이라기보다는, [Brownlee의 블로그](https://machinelearningmastery.com/the-bahdanau-attention-mechanism/)의 설명을 내가 이해한 방식대로 옮긴 것이다.
그런데 이 Brownlee의 설명은 정확한 것 같다.
논문하고 비교해봐도 크게 다르지 않은 것 같이 보인다.
특히, [유원준님의 wikidocs 페이지](https://wikidocs.net/73161)에는, 쉽게 설명하는 것을 목적으로 써놓은 것 같기는 하지만, 논문과 비교했을 때 notation 같은게 좀 달라보인다.
그래도 유원준님의 설명에서도 가지고 올 만한 것들은 가져오려고 했다.
일단, 아까 그림을 다시 가져와보자.

vanilla seq2seq와 이 모델의 차이는, context vector가 어떻게 정의되느냐 하는 점이다.
context vector는 attention을 사용하여 만들어진다.
attention을 사용하는 것이라면, query, key, value가 각각 무엇인지 알아야 할 것이다.
아까는 이것이 명확하지 않았는데, 유원준님의 글을 통해 보면 다음과 같다.
 - query : $s_{t-1}$
 - key : $h_1$, $\cdots$, $h_N$
 - value : $h_1$, $\cdots$, $h_N$
 
이 모델에서의 관건은, $s_1$, $\cdots$, $s_{t-1}$이 주어진 상태에서, $s_t$를 어떻게 정할 수 있겠느냐, 하는 문제이다.
이것이 정의되고 나면 당연히 모든 $t$에 대하여 $s_t$가 inductive하게 (차례로) 결정될 것이다.

그리고 사실 중요한 계산은 위의 식에 다 써놓았다.
그 과정을 다시 한 번 적어보자.
$t$는 고정되어 있다.
일단, (대개의 attention 모델들이 그렇듯이) query($s_{t-1}$)와 key($h_1$, $\cdots$, $h_N$)들을 가지고 어떤 종류의 계산을 한다.
\begin{align*}e_{t,i}&=a(s_{t-1},h_i)\\
&=v^T\tanh\left(W[h_i;s_{t-1}]\right)\quad(i=1,\cdots, N)
\end{align*}

그것을 normalize하여 (softmax) 확률과 비슷한 값 $\alpha_{t,i}$으로 만든다.
그 값은, 위의 그림에 적혀있는 attentnion weight의 역할을 한다.
이 attention weight들은 value인 $h_1$, $\cdots$, $h_N$에 적당히 분배되어 linear combination을 만든 뒤 context vector $c_t$에 저장된다.
그러니까, context vector는 t에만 의존한다.
이 context vector $c_t$가 $h_t$와 어떻게 결합해서 결과를 내는지에 대해서는 brownlee의 글에는 잘 나타나있지 않고, 논문에도 잘 안보인다.
다만 유원준님의 글에는 이렇게 설명이 되어 있다.
context vector $c_t$는 embedding layer를 거치고 난 벡터 $u_{t-1}$와 concatenate하여  그것을 입력값으로 하여 LSTM을 돌린다는 것이다.

다음과 같은 그림으로 설명하고 있다.



이후에는 익히 아는 대로 진행된다. (하지만 정확히는 알지 못하겠다. 정확히 어떤 식으로 softmax가 적용되는지 말이다.) 대략적으로 말하면, 2000개 노드의 softmax 레이어를 넣고 여기에 argmax를 취해서 2000차원의 one-hot vector를 만들고 그것을 통해 단어로 변환하는 것이다.
하지만 이것은 한 단어에 대해 말한 것이고, 실제로는 8개의 hidden state에 대하여 (n_timesteps=8) 그것들을 단어로 변환시킨다.

정확히 한 번 써보자. hidden state vector들이 모두 출력되었다면, 그것은 256차원의 벡터를 표현하는 8개의 노드가 생성되었다는 것을 뜻한다.
다시 말해, 하나의 노드에는 하나의 실수가 대응되는 게 아니라 여러개의 실수, 256개의 실수, 즉 256차원의 벡터가 대응된다.
이것은 8 vectors of dimension 256이다. 이 상태에서 fully connected layer를 취하는데, 각 노드에 대하여 독립적으로 fullyconnected layer를 만들든지, 아니면 같은 weight set을 사용해서 fully connected layer를 만들든지 하게 될 것 같다.
즉 input layer가 256차원이고, output layer가 2000차원인 fully connected layer를 생각하는 것이다.
그리고 activation function을 softmax로 두면 되겠지만, 굳이 softmax로 두지 않아도 괜찮다.
어차피 필요한 것은 2000개의 숫자 중에 뭐가 제일 큰지 하는 것이기 때문이다.
그러면 각 8개의 벡터들은 (원래 256차원의 벡터였다가, 다음에는 2000차원의 벡터가 된 그 벡터들은) one-hot vector로 바뀐다.
이 one-hot vector들은 lookup table을 통해, 바로 단어로 변환될 수 있다. 이렇게 모델의 뒷부분이 구성되었을 것이라고 생각된다.
아래에 조악하게나마 그림을 그려봤다.

```python
# 코드를 내가 이해할 수 있을까? 못할 것 같지만 한 번 해보자.
#
# decoder_outputs # s_t or the enumeration of s_t where t ranges from 0 to 7
# [None] sentences of 8 words of 256 dimensional vectors (None, None=8, 256)
# encoder_outputs # h_i or the enumeration of s_i where i ranges from 0 to 49
# [None] sentences of 50 words of 256 dimensional vectors (None, 50, 256)
# attn_out : (None, None=8, 256)
# attn_out은, 위의 두 텐서를 모든 i=0,...49에 대하여 계산하여 얻은 c_t라고 생각된다.
# c_t는 h_i(encoder_outputs)들을 weighted sum한 값이니, encoder_outputs과 shape이 정확히 일치한다.
# 이 c_t는 각각의 decoder timestep t에 대하여 계산된다. 정확히 그런 점에서도 attn_out의 shape이 (None, None=8, 256)라는 게 말이 된다.
# 
# 정리해보자
# decoder_outputs : s_t
# encoder_outputs : h_i
# atten_out : c_t
#
# 이후에는 조금 간단해보인다. Concatenate라는 레이어를 사용하는데, 이것은 단순한 concatenation이라고 생각하면 될 것 같다.
# 즉 s_t와 c_t를 concat한다.
# 하지만 정확하게는, 아직 s_{t-1}까지밖에 없는 상태이고 s_t는 없는 상태이다. 지금 하고 싶은 것이 s_t를 구하려는 것이다.
# 여하튼 s_t와 c_t를 concat한 것에 FC layer를 넣는데 이때 node의 개수를 tar_vocab=2000으로 잡고, activation을 softmax로 한다.
#
# decoder_softmax_outputs의 shape은 (None, None=8, 2000)이다. 정확히 잘 되었다.
#
# 모델을 새로 구성한다. 기존에 정의한 모델은 더이상 없다.
from tensorflow.keras.layers import AdditiveAttention

# 어텐션 층(어텐션 함수)
attn_layer = AdditiveAttention(name='attention_layer')

# 인코더와 디코더의 모든 time step의 hidden state를 어텐션 층에 전달하고 결과를 리턴
attn_out = attn_layer([decoder_outputs, encoder_outputs])

# 어텐션의 결과와 디코더의 hidden state들을 연결
decoder_concat_input = Concatenate(axis=-1, name='concat_layer')([decoder_outputs, attn_out])

# 디코더의 출력층
decoder_softmax_layer = Dense(tar_vocab, activation='softmax')
decoder_softmax_outputs = decoder_softmax_layer(decoder_concat_input)

# 모델 정의
model = Model([encoder_inputs, decoder_inputs], decoder_softmax_outputs)
model.summary()
```

```python
# 이것은 Bahdanau 스타일의 architecture가 아니다. Bahdanau 스타일에서는 bidirectional LSTM을 썼다.
# 아마 프로젝트에서는 bidrectional을 쓰는 것 같은데 맞는지 모르겠다.
```

## 2.3. 모델 학습

```python
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy')
es = EarlyStopping(monitor='val_loss', patience=2, verbose=1)
history = model.fit(x=[encoder_input_train, decoder_input_train], y=decoder_target_train, \
          validation_data=([encoder_input_test, decoder_input_test], decoder_target_test), \
          batch_size=256, callbacks=[es], epochs=50)
```

## 2.4. 추론 모델 구성

```python
src_index_to_word = src_tokenizer.index_word # 원문 단어 집합에서 정수 -> 단어를 얻음
tar_word_to_index = tar_tokenizer.word_index # 요약 단어 집합에서 단어 -> 정수를 얻음
tar_index_to_word = tar_tokenizer.index_word # 요약 단어 집합에서 정수 -> 단어를 얻음

print('=3')
```

```python
# 인코더 설계
encoder_model = Model(inputs=encoder_inputs, outputs=[encoder_outputs, state_h, state_c])

# 이전 시점의 상태들을 저장하는 텐서
decoder_state_input_h = Input(shape=(hidden_size,))
decoder_state_input_c = Input(shape=(hidden_size,))

dec_emb2 = dec_emb_layer(decoder_inputs)

# 문장의 다음 단어를 예측하기 위해서 초기 상태(initial_state)를 이전 시점의 상태로 사용. 이는 뒤의 함수 decode_sequence()에 구현
# 훈련 과정에서와 달리 LSTM의 리턴하는 은닉 상태와 셀 상태인 state_h와 state_c를 버리지 않음.
decoder_outputs2, state_h2, state_c2 = decoder_lstm(dec_emb2, initial_state=[decoder_state_input_h, decoder_state_input_c])

print('=3')
```

```python
# 어텐션 함수
decoder_hidden_state_input = Input(shape=(text_max_len, hidden_size))
attn_out_inf = attn_layer([decoder_outputs2, decoder_hidden_state_input])
decoder_inf_concat = Concatenate(axis=-1, name='concat')([decoder_outputs2, attn_out_inf])

# 디코더의 출력층
decoder_outputs2 = decoder_softmax_layer(decoder_inf_concat) 

# 최종 디코더 모델
decoder_model = Model(
    [decoder_inputs] + [decoder_hidden_state_input,decoder_state_input_h, decoder_state_input_c],
    [decoder_outputs2] + [state_h2, state_c2])

print('=3')
```

```python
def decode_sequence(input_seq):
    # 입력으로부터 인코더의 상태를 얻음
    e_out, e_h, e_c = encoder_model.predict(input_seq)

     # <SOS>에 해당하는 토큰 생성
    target_seq = np.zeros((1,1))
    target_seq[0, 0] = tar_word_to_index['sostoken']

    stop_condition = False
    decoded_sentence = ''
    while not stop_condition: # stop_condition이 True가 될 때까지 루프 반복

        output_tokens, h, c = decoder_model.predict([target_seq] + [e_out, e_h, e_c])
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_token = tar_index_to_word[sampled_token_index]

        if (sampled_token!='eostoken'):
            decoded_sentence += ' '+sampled_token

        #  <eos>에 도달하거나 최대 길이를 넘으면 중단.
        if (sampled_token == 'eostoken'  or len(decoded_sentence.split()) >= (headlines_max_len-1)):
            stop_condition = True

        # 길이가 1인 타겟 시퀀스를 업데이트
        target_seq = np.zeros((1,1))
        target_seq[0, 0] = sampled_token_index

        # 상태를 업데이트 합니다.
        e_h, e_c = h, c

    return decoded_sentence
print('=3')
```

# 3. 결과

```python
# 원문의 정수 시퀀스를 텍스트 시퀀스로 변환
def seq2text(input_seq):
    temp=''
    for i in input_seq:
        if (i!=0):
            temp = temp + src_index_to_word[i]+' '
    return temp

# 요약문의 정수 시퀀스를 텍스트 시퀀스로 변환
def seq2summary(input_seq):
    temp=''
    for i in input_seq:
        if ((i!=0 and i!=tar_word_to_index['sostoken']) and i!=tar_word_to_index['eostoken']):
            temp = temp + tar_index_to_word[i] + ' '
    return temp

print('=3')
```

```python
for i in range(50, 100):
    print("원문 :", seq2text(encoder_input_test[i]))
    print("실제 요약 :", seq2summary(decoder_input_test[i]))
    print("예측 요약 :", decode_sequence(encoder_input_test[i].reshape(1, text_max_len)))
    print("\n")
```

2.
구체적으로 이야기해보면, seq2seq 모델 하나를 구현했는데, 그것이 이전에 구성한 seq2seq 모델과는 attention을 도입했다는 점에서 다릅니다.
기존의 seq2seq는 encoder의 마지막 hidden state vector를 context vector로 두었는데, 이번에는 attention의 query($s_{t-1}$)와 key($h_i$)로 만든 weight들을 가지고, value($h_i$)들의 weighted sum을 context로 정의했습니다.
이 context vector는 decoder의 timestep에 의존한다는 것, 다시 말해, decoder의 각 timestep마다 서로 다른 context vector를 사용한다는 점이 기존과 다릅니다.
요약하면, (1) context vector가 정교하게 구성된 weighted sum이라는 점과 (2) decoder의 timestep에 따라 달라진다는 점에서 context vector가 기존과 상당히 달라졌습니다.
이렇게 함으로써, 각각의 decoder timestep에 맞는 context vector를 가져올 수 있다는 점이 있었습니다.
그러니까, headlines 혹은 summary의 첫번째 단어와 두번째 단어, 그리고 세 번째 단어를 generate하는 데 있어서 서로 다른 문맥을 사용하게 된 것이고, 그렇게 서로 다른 문맥을 사용하는 것이 더 정교하고 옳은 방법일 것입니다.
그 점에서 attention을 사용한 seq2seq는 기존의 vanilla seq2seq에 비해 발전한 방법입니다.  
3. 한편, 실제 모델이 요약해낸 문장들이 '잘 요약했는지'는 잘 모르겠습니다. 아마도 잘 요약하지 못한 것들이 더 많아보이기도 합니다.  
4. 또 한가지 언급할만한 점은, 기존의 노드에 담겨있는 코드와 비교했을 때 제 코드에서 다르게 한 점이 몇가지 있다는 점입니다.
기존 노드에서는 `text_max_len`만을 지정했는데, 이것은 전처리시에 문장의 길이 (단어의 수)의 상한(upper bound)을 제한했었던 것입니다.
그때는 그럴 만했던 것이, 문장의 길이의 분포를 봤을 때, 문장이 상당히 긴 것들이 존재했으며, 그것이 아주 sparse하게 존재했기 때문입니다.
그러니까, 적당한 threshold, 즉 upper bound를 잡아서 자르는 것이 맞았습니다. 하지만 뉴스 요약의 경우, 문장의 길이의 분포가 종모양의 정규분포 곡선 모양을 띄었다는 점에서 달랐습니다.
사실 너무 예쁜 모양의 정규분포였기에 굳이 outlier들을 없애지 않는다고 해도 좋은 성능이 나오리라고 생각되었지만, 기왕 `below_threshold_len`함수도 있는데 안쓰기가 아까워서 썼습니다.
다만, 기존에는 upper bound만을 고려해 잘랐다면 이번에는 lower bound까지 고려해서 잘랐습니다.
(사실 그렇게 할 거면 함수명을 바꿨어야 할 것 같은데, 바꾸지는 않았습니다.)
